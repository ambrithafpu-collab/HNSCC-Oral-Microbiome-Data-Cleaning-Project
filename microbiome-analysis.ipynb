```python
import pandas as pd

data = pd.read_csv("data.csv")

print("Original Data:")
data

clean_data = data.dropna()

print("\nCleaned Data:")
clean_data 
```

```python
import matplotlib.pyplot as plt

# Plot bar graph
plt.bar(clean_data["Bacteria"], clean_data["Count"])

plt.xlabel("Bacteria")
plt.ylabel("Count")
plt.title("Microbiome Data Analysis")

plt.xticks(rotation=45)
plt.show()
```
